```python
import pandas as pd
from sportsdataverse.nba import load_nba_player_boxscore

pd.set_option("display.max_columns", 100)
```

```python
box_df = load_nba_player_boxscore(seasons=2022)\
    .to_pandas(date_as_object=False)
```

```python
lebron = box_df.query("athlete_display_name == 'LeBron James'")
```

```python
lebron
```

```python
lebron.info()
```

```python
# save lebron's boxscores to csv
lebron.to_csv("lebron.csv", index=False)
```

```python
# read in lebron's boxscores
lebron = pd.read_csv("lebron.csv")
```

```python
lebron.head()
```

```python
lebron.tail()
```

```python
lebron.describe()
```

```python
# get max number of minutes lebron played in a game this season
lebron["minutes"].max()
```

```python
# filter to only include games where lebron played 42+ minutes
lebron.query("minutes >= 42")
```

```python
# filter to only include games where lebron played 42+ minutes AND attempted less than 7 free throws
lebron.query("minutes >= 42 & free_throws_attempted < 7")
```

```python
# find the game where lebron played the least amount of minutes
lebron.query("minutes == minutes.min()")
```

```python
# total points scored this season
lebron.points.sum()
```

```python
# Distribution of Lebron's points
lebron.points.value_counts().sort_index()
```

```python
# Plot Distribution of Lebron's points
lebron.points.value_counts().sort_index().plot();
```
